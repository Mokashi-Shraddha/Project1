```python
import unittest
import pytest
from modules.evaluate_model import evaluate_sentiment_model, evaluate_sarcasm_model

def test_evaluate_sentiment_model():
    # Assuming a sentiment model and test data are available
    sentiment_model = load_sentiment_model()
    test_data = [("I love this movie!", "positive"), ("I hate this movie!", "negative")]
    accuracy = evaluate_sentiment_model(sentiment_model, test_data)
    assert 0 <= accuracy <= 1

def test_evaluate_sarcasm_model():
    # Assuming a sarcasm model and test data are available
    sarcasm_model = load_sarcasm_model()
    test_data = [("Oh, that's just what I needed!", "sarcastic"), ("This is great!", "non-sarcastic")]
    accuracy = evaluate_sarcasm_model(sarcasm_model, test_data)
    assert 0 <= accuracy <= 1
```
